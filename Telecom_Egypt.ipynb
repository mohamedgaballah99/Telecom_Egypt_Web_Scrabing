```python
import requests 
import csv
from bs4 import BeautifulSoup
```

```python
search=input("Choose Your Category From This List..?[5G-Devices-Mobile , Accessories , 4G-Routers , Routers , USB-Modems , Mobile-Phones , Fixed-Landline-Phones]")
link=f'https://te.eg/wps/portal/te/Personal/Devices/{search}'
requests.get(link)
```

```python
we_list=[]
def main(link):
    page=requests.get(link)
    soup=BeautifulSoup(page.text,'html.parser')
    
    item_header=soup.find_all('a',{'class','item active'})
    #print(item_header)
    devices_details=soup.find_all('div',{'class','compound-title'})

    for item in item_header:
        header=item.find('h4').text if item else ''
        print(header)
    for device in devices_details:
        titel= device.find('div',{'class','pro-card-head'}).text.strip() if device else ''
        price=device.find('span',{'class','price'}).text.strip() if device else ''
        company=device.find('span',{'class','manufactur'}).text.strip() if device else ''
        print(titel, ' ', price, ' ', company)

        we_list.append({
             'Title':titel,
             'Company':company,
             'Price':price
            })
    
```

```python
def printing():
    sheet=we_list[0].keys()
    path=f'F:\Data-Analysis-Course\Python\WE/{search}.csv'
    with open(path,'w',encoding='utf-8',newline='') as file:
        writer=csv.DictWriter(file,sheet)
        writer.writeheader()
        writer.writerows(we_list)
        print('DONE')
```

```python
main(link)
printing()
```
